## Creating New Variables

```python
import pandas as pd
```

```python
data = pd.read_csv('data/aluguel_residencial.csv', sep = ';')
```

```python
data.head()
```

```python
data['Valor Bruto'] = data['Valor'] + data['Condominio'] + data['IPTU']
```

```python
data.head()
```

```python
data['Valor m²'] = data['Valor'] / data['Area']
```

```python
data.head()
```

```python
data['Valor m²'] = data['Valor m²'].round(2)
```

```python
data.head()
```

```python
data['Valor Bruto m²'] = (data['Valor Bruto']/data['Area']).round(2)
```

```python
data.head()
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
data['Tipo Agregado'] = data['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento') # O apply permite a aplicação de uma função a cada registro.
```

```python
data
```
